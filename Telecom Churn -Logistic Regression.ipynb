```python
#Telecorm Churn Notebook
#import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#get and combine all data sets
cust_data = pd.read_csv('customer_data.csv')

cust_data.info()
```

```python
cust_data.info()
```

```python
internet_data = pd.read_csv('internet_data.csv')
internet_data
```

```python
churn_data = pd.read_csv('churn_data.csv')
churn_data
```

```python
#merge the 3 data sets on customer_id
master_data_1 = cust_data.merge(internet_data,how = 'inner',right_on ='customerID',left_on ='customerID')
master_data_1
```

```python
master_data = master_data_1.merge(churn_data,how = 'inner',right_on ='customerID',left_on ='customerID')
master_data.info()
```

```python
master_data.info()
```

```python
#The varaible was imported as a string we need to convert it to float
master_data['TotalCharges'] = master_data['TotalCharges'].convert_objects(convert_numeric=True)
```

```python
#Chcking missing data
master_data.isnull().sum()
```

```python
#we can see that all columns are present so now analysind dataframe for qulity
master_data.head()
```

```python
#Converting categorical varibales
#Binary conversion of yes no
varlist =['PhoneService','PaperlessBilling','Churn','Partner','Dependents']

def binary_map(col):
    return col.map({"Yes":1,"No":0})

master_data[varlist] = master_data[varlist].apply(binary_map)
master_data['gender'] = master_data['gender'].map({"Male":1,"Female":0})
master_data
```

```python
#creation of dummy variables
def create_dummies(df,col):
    dummies = pd.get_dummies(df[col],prefix=col,drop_first =True)
    df = pd.concat([df,dummies],axis =1)
    df.drop(columns = col,inplace = True)
    return df
master_data.nunique()

master_data = create_dummies(master_data,'OnlineSecurity') 
master_data = create_dummies(master_data,'OnlineBackup')
master_data = create_dummies(master_data,'DeviceProtection')
master_data = create_dummies(master_data,'TechSupport') 
master_data = create_dummies(master_data,'StreamingTV') 
master_data = create_dummies(master_data,'StreamingMovies')
master_data = create_dummies(master_data,'MultipleLines') 
master_data = create_dummies(master_data,'InternetService')
master_data = create_dummies(master_data,'Contract') 
master_data = create_dummies(master_data,'PaymentMethod')



```
